# IR-SIM Quick Start

Run a differential-drive robot navigating to a goal — entirely in your browser, no install needed.

This notebook: installs IR-SIM, writes a small YAML scenario, runs the simulation headless, and shows the result as an animation.

```python
!pip install -q ir-sim
```

Define the scenario in a YAML file. A 10×10 world with one differential-drive robot that dashes from `[1, 1]` to its goal at `[9, 9]`.

```python
%%writefile robot_world.yaml
world:
  height: 10
  width: 10
  step_time: 0.1
  sample_time: 0.1
  offset: [0, 0]

robot:
  kinematics: {name: 'diff'}
  shape: {name: 'circle', radius: 0.2}
  state: [1, 1, 0]
  goal: [9, 9, 0]
  behavior: {name: 'dash'}
  color: 'g'
  plot:
    show_trajectory: True
    show_goal: True
```

Run the simulation. We use a non-interactive Matplotlib backend and `display=False` because the notebook has no on-screen window; `save_ani=True` records the run, and `env.end()` writes it to an animated GIF.

```python
import matplotlib
matplotlib.use("Agg")
import irsim

env = irsim.make("robot_world.yaml", save_ani=True, display=False)

for i in range(300):
    env.step()
    env.render(0.05)
    if env.done():
        break

env.end(ending_time=3)
```

Show the recorded animation.

```python
from IPython.display import Image

Image("animation/animation_robot_world.gif")
```
